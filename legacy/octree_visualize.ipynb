```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
filename = "temp.dat"

with open(filename, 'r') as file:
    squares = []
    for line in file:
        # Split the line into parts and convert them to integers
        numbers = [int(num) for num in line.split()]
        squares.append(numbers)
```

```python
# Create a new figure
fig, ax = plt.subplots()

# Add each square to the plot
for square in squares:
    lower_left_x, lower_left_y, upper_right_x, upper_right_y = square
    width = upper_right_x - lower_left_x
    height = upper_right_y - lower_left_y
    rect = patches.Rectangle((lower_left_x, lower_left_y), width, height, linewidth=1, edgecolor='black', facecolor='none')
    ax.add_patch(rect)

cntr = (1<<12)*0.5*10
circle_center = (cntr, cntr)  # Center of the circle (x, y)
circle_radius = (1<<12)*0.5*10*0.7      # Radius of the circle
circle = plt.Circle(circle_center, circle_radius, color='blue', fill=False)
ax.add_patch(circle)

# Set the limits of the plot
ax.set_xlim([0, 65536])
ax.set_ylim([0, 65536])
ax.set_aspect('equal', adjustable='box')

# Show the plot
plt.show()
```
